# Benchmark 01: Single vs Concurrent Request Latency

**Goal:** Measure how response time changes when we go from 1 to N concurrent requests.

**Setup:** Ollama running locally with llama3.2

**What we're measuring:**
- TTFT (Time to First Token) — how long until the first word appears
- Total response time per request
- How latency degrades under concurrent load

**Why this matters:** This is the most basic inference serving tradeoff — latency vs throughput. Understanding this from real numbers is foundational to everything in vLLM, PagedAttention, and batching strategies.

## Setup — make sure Ollama is running

Before running this notebook:
```bash
ollama serve
ollama pull llama3.2
```

````python
import requests
import time
import json
import threading
import statistics

OLLAMA_URL = "http://localhost:11434/api/generate"
MODEL = "llama3.2"
PROMPT = "Explain what a KV cache is in two sentences."

def single_request(prompt=PROMPT, stream=False):
    """Send one request and measure total time and TTFT."""
    payload = {
        "model": MODEL,
        "prompt": prompt,
        "stream": stream
    }
    
    start = time.time()
    
    if stream:
        # Streaming: capture time to first token
        first_token_time = None
        full_response = ""
        with requests.post(OLLAMA_URL, json=payload, stream=True) as r:
            for line in r.iter_lines():
                if line:
                    data = json.loads(line)
                    if first_token_time is None and data.get("response"):
                        first_token_time = time.time() - start
                    full_response += data.get("response", "")
                    if data.get("done"):
                        break
        total_time = time.time() - start
        return {
            "ttft": first_token_time,
            "total_time": total_time,
            "response": full_response[:100]
        }
    else:
        r = requests.post(OLLAMA_URL, json=payload)
        total_time = time.time() - start
        return {
            "ttft": total_time,  # non-streaming: entire response at once
            "total_time": total_time,
            "response": r.json().get("response", "")[:100]
        }

print("Functions loaded. Ready to benchmark.")
````

## Experiment 1: Single request baseline

````python
print("Running single request baseline...")
print("-" * 40)

# Run 3 times and average to account for variance
single_results = []
for i in range(3):
    result = single_request()
    single_results.append(result["total_time"])
    print(f"  Run {i+1}: {result['total_time']:.2f}s")

avg_single = statistics.mean(single_results)
print(f"\nAverage single request time: {avg_single:.2f}s")
print(f"Response preview: {result['response'][:80]}")
````

## Experiment 2: Concurrent requests

We send N requests at the same time using threads.
Each thread fires independently — simulating N users hitting the server simultaneously.

**What to watch:** Does per-request latency stay flat, or does it increase as concurrency grows?

````python
def concurrent_requests(n):
    """Fire n requests simultaneously using threads."""
    results = [None] * n
    threads = []
    
    def run(i):
        results[i] = single_request()
    
    # Start all threads at the same time
    wall_start = time.time()
    for i in range(n):
        t = threading.Thread(target=run, args=(i,))
        threads.append(t)
    
    for t in threads:
        t.start()
    
    for t in threads:
        t.join()
    
    wall_time = time.time() - wall_start
    per_request_times = [r["total_time"] for r in results]
    
    return {
        "n": n,
        "wall_time": wall_time,
        "per_request_avg": statistics.mean(per_request_times),
        "per_request_max": max(per_request_times),
        "per_request_min": min(per_request_times),
        "all_times": per_request_times
    }

print("Concurrent request function ready.")
````

````python
print("Running concurrency sweep: 1, 2, 4 concurrent requests")
print("(This will take a few minutes — each set waits for all requests to finish)")
print("=" * 60)

concurrency_results = []

for n in [1, 2, 4]:
    print(f"\nRunning {n} concurrent request(s)...")
    result = concurrent_requests(n)
    concurrency_results.append(result)
    print(f"  Wall time (total):        {result['wall_time']:.2f}s")
    print(f"  Per-request avg latency:  {result['per_request_avg']:.2f}s")
    print(f"  Per-request max latency:  {result['per_request_max']:.2f}s")
    print(f"  Individual times: {[f'{t:.2f}s' for t in result['all_times']]}")
````

## Results Summary

Fill this in after running the cells above.

````python
print("\n" + "=" * 60)
print("BENCHMARK RESULTS SUMMARY")
print("=" * 60)
print(f"Model:  {MODEL}")
print(f"Setup:  Ollama local")
print(f"Prompt: '{PROMPT}'")
print()
print(f"{'Concurrency':<15} {'Wall time':<15} {'Avg per-req':<15} {'Max per-req':<15}")
print("-" * 60)

for r in concurrency_results:
    print(f"{r['n']:<15} {r['wall_time']:<15.2f} {r['per_request_avg']:<15.2f} {r['per_request_max']:<15.2f}")

print()
print("KEY OBSERVATION:")
if len(concurrency_results) >= 3:
    slowdown = concurrency_results[2]['per_request_avg'] / concurrency_results[0]['per_request_avg']
    print(f"Per-request latency at 4 concurrent vs 1: {slowdown:.1f}x slower")
    print()
    print("This is the core latency/throughput tradeoff in inference serving.")
    print("Ollama serves requests sequentially — each request waits for the previous.")
    print("vLLM's continuous batching solves this by interleaving token generation.")
````

## What these numbers mean

**If per-request latency stays roughly flat** as concurrency increases:
- The server is handling requests in parallel
- Good news for throughput

**If per-request latency increases linearly** (2x at 2 concurrent, 4x at 4 concurrent):
- Requests are being queued and served one at a time
- This is Ollama's default behavior — it's a local dev tool, not a production server
- This is exactly the problem vLLM's continuous batching solves

**The insight:** 
The roofline model told us inference is memory-bandwidth-bound. 
These numbers show what that means in practice for a real server under load.
A single request gets the full GPU. Four concurrent requests share it — or wait in line.
